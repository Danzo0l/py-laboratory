```python
import numpy as np
import matplotlib.pyplot as plt

# Заданные значения случайной величины и их вероятности
x_values = [-73.4, -70.7, -51.5, -43.9, 13.3, 73.0, 73.8]
p_values = [0.241, 0.023, 0.166, 0.078, 0.272, 0.192, 0.028]

# Генерация выборки из дискретной случайной величины
data_discrete = np.random.choice(x_values, size=500, p=p_values)

# Вывод первых 30 величин
print("Первые 30 величин выборки:")
print(data_discrete[:30])

# Эмпирические оценки M и D
mean_empirical = np.mean(data_discrete)
var_empirical = np.var(data_discrete)

# Теоретические оценки M и D
mean_theoretical = np.sum(np.array(x_values) * np.array(p_values))
var_theoretical = np.sum(np.array([(x - mean_theoretical) ** 2 for x in x_values]) * np.array(p_values))

# Построение гистограммы эмпирических вероятностей
plt.bar(x_values, p_values, color='r', alpha=0.6, label='Теоретические вероятности', width=4)

# Добавление гистограммы для отображения выборки
plt.hist(data_discrete, bins=len(x_values), density=True, alpha=0.6, color='b', label='Эмпирические вероятности', align='mid', rwidth=0.8)

plt.xlabel('Значение случайной величины')
plt.ylabel('Вероятность')
plt.title('Сравнение эмпирических и теоретических вероятностей')
plt.legend()
plt.grid(True)
plt.show()

# Вывод оценок
print("Эмпирическая оценка M:", mean_empirical)
print("Теоретическая оценка M:", mean_theoretical)
print("Эмпирическая оценка D:", var_empirical)
print("Теоретическая оценка D:", var_theoretical)
```
