```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создаем данные
x = np.linspace(0, 1, 100)  # Ось x от 0 до 1
y = np.linspace(0, 1, 100)  # Ось y от 0 до 1
X, Y = np.meshgrid(x, y)
Z = np.where((X <= 1) & (Y <= 1) & (X >= Y), 2, np.nan)  # Ограничиваем область

# Создаем 3D-график
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Построение поверхности
ax.plot_surface(X, Y, Z, alpha=0.1, rstride=100, cstride=100, cmap='gray')

# Добавляем графики y=x, x=1, y=1 в плоскости z=0 и z=2
ax.plot(x, x, 0, color='gray', linewidth=2)
ax.plot([1, 1], [0, 1], 0, color='gray', linewidth=2)
ax.plot([0, 1], [1, 1], 0, color='gray', linewidth=2)

ax.plot(x, x, 2, color='gray', linewidth=2)
ax.plot([1, 1], [0, 1], 2, color='gray', linewidth=2)
ax.plot([0, 1], [1, 1], 2, color='gray', linewidth=2)

# Настройка осей
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])


# Отображаем график
plt.show()
```
