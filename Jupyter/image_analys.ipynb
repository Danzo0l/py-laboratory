```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math

def count_different_pixels(file1, file2):
    img1 = np.array(Image.open(file1))
    img2 = np.array(Image.open(file2))

    different_pixels = np.sum(img1 != img2)

    return different_pixels

def calculate_correlation(file1, file2):
    img1 = np.array(Image.open(file1))
    img2 = np.array(Image.open(file2))

    flat_img1 = img1.flatten()
    flat_img2 = img2.flatten()
    correlation_coefficient = np.corrcoef(flat_img1, flat_img2)[0, 1]
    return correlation_coefficient
```

```python
if __name__ == '__main__':
    original_file = 'C:\\Users\\dryag\\Downloads\\input.bmp'
    encrypted_file = 'C:\\Users\\dryag\\Downloads\\out_image.bmp'
    correlation = calculate_correlation(original_file, encrypted_file)
    print(f"correlation_coefficient: {math.fabs(correlation)}")
    print("The correlation coefficient for plain image (Ape.bmp) is 0.874144.")
    print("The correlation coefficient for the cipher image encrypted using original CAST-128 is 0.016693.")
    print("The correlation coefficient for the cipher image encrypted using modified CAST-128 is 0.012245.")

    img1 = Image.open(original_file)
    img2 = Image.open(encrypted_file)

    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.title('Source image')
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.title('Decrypted image')

    plt.show()
```
