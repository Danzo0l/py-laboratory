```python
import numpy as np
from matplotlib import pyplot as plt

from mMTC.ZC_sequence import Z_r, Z_p_r

# Conventional Preamble

NZC = 23
r = 7
p = 3

n_values = np.arange(0, NZC)
z_values = Z_r(r, NZC)
z_p_values = Z_p_r(p, z_values)

# Plot ZC Sequence
plt.figure(figsize=(4, 3))
plt.plot(n_values, np.real(z_values), label='Real part')
plt.plot(n_values, np.imag(z_values), label='Imaginary part')
plt.xlabel('n')
plt.ylabel('z_r[n]')
plt.title('ZC Sequence')
plt.legend(loc='lower right')
plt.show()

# Plot complex numbers on the complex plane
plt.figure(figsize=(4.4, 2.2))
plt.subplot(1, 2, 1)
plt.scatter(np.real(z_values), np.imag(z_values), color='red', label='Complex numbers', marker='o')
plt.title('ZC')
plt.axis('equal')
plt.subplot(1, 2, 2)
plt.scatter(np.real(z_p_values), np.imag(z_p_values), color='red', label='Complex numbers', marker='o')
plt.title('ZC (shifted)')
plt.axis('equal')
plt.show()

print(np.dot(z_values, z_p_values))


ZC = z_values
ZC_shifted = z_p_values

# Print the arrays as tables
ZC = z_values
ZC_shifted = z_p_values

# Print the arrays as tables side by side
print('ZC\t\t\tZC (shifted)')
for num1, num2 in zip(ZC, ZC_shifted):
    rounded_num1 = np.round(num1, decimals=4)
    rounded_num2 = np.round(num2, decimals=4)
    print('{:<23} {:<23}'.format(rounded_num1, rounded_num2))
```
