```python
import numpy as np
import matplotlib.pyplot as plt

# Параметр экспоненциального распределения
lambd = 0.5

# Генерация выборки из экспоненциального распределения
data_exp = np.random.exponential(scale=1/lambd, size=1000)

# Построение гистограммы
plt.hist(data_exp, bins=30, density=True, alpha=0.6, color='b')

plt.xlabel('Значение случайной величины')
plt.ylabel('Плотность вероятности')
plt.title('Гистограмма экспоненциального распределения')
plt.grid(True)
plt.show()

# Эмпирические значения
mean_empirical_exp = np.mean(data_exp)
var_empirical_exp = np.var(data_exp)

# Теоретические значения
mean_theoretical_exp = 1 / lambd
var_theoretical_exp = 1 / lambd ** 2

print("Экспоненциальное распределение:")
print("Теоретическое среднее:", mean_theoretical_exp)
print("Эмпирическое среднее:", mean_empirical_exp)
print("Теоретическая дисперсия:", var_theoretical_exp)
print("Эмпирическая дисперсия:", var_empirical_exp)

```

```python
# Генерация выборки из равномерного распределения
data_unif = np.random.uniform(0, 1, 1000)

# Построение гистограммы
plt.hist(data_unif, bins=30, density=True, alpha=0.6, color='r')

plt.xlabel('Значение случайной величины')
plt.ylabel('Плотность вероятности')
plt.title('Гистограмма равномерного распределения')
plt.grid(True)
plt.show()

# Параметры равномерного распределения
a_unif = 0
b_unif = 1

# Эмпирические значения
mean_empirical_unif = np.mean(data_unif)
var_empirical_unif = np.var(data_unif)

# Теоретические значения
mean_theoretical_unif = (a_unif + b_unif) / 2
var_theoretical_unif = (b_unif - a_unif) ** 2 / 12

print("Равномерное распределение:")
print("Теоретическое среднее:", mean_theoretical_unif)
print("Эмпирическое среднее:", mean_empirical_unif)
print("Теоретическая дисперсия:", var_theoretical_unif)
print("Эмпирическая дисперсия:", var_empirical_unif)

```

```python
from scipy.stats import gamma

# Параметры распределения Эрланга порядка K
k_erlang = 2
theta_erlang = 1.5

# Генерация выборки из распределения Эрланга
data_erlang = np.random.gamma(k_erlang, theta_erlang, 1000)

# Построение гистограммы
plt.hist(data_erlang, bins=30, density=True, alpha=0.6, color='g')

plt.xlabel('Значение случайной величины')
plt.ylabel('Плотность вероятности')
plt.title('Гистограмма распределения Эрланга порядка {}'.format(k_erlang))
plt.grid(True)
plt.show()

# Параметры распределения Эрланга
k_erlang = 2
theta_erlang = 1.5

# Эмпирические значения
mean_empirical_erlang = np.mean(data_erlang)
var_empirical_erlang = np.var(data_erlang)

# Теоретические значения
mean_theoretical_erlang = k_erlang * theta_erlang
var_theoretical_erlang = k_erlang * theta_erlang ** 2

print("Распределение Эрланга порядка", k_erlang)
print("Теоретическое среднее:", mean_theoretical_erlang)
print("Эмпирическое среднее:", mean_empirical_erlang)
print("Теоретическая дисперсия:", var_theoretical_erlang)
print("Эмпирическая дисперсия:", var_empirical_erlang)

```

```python
# Параметры нормального распределения
mu_norm = 0
sigma_norm = 1

# Генерация выборки из нормального распределения
data_norm = np.random.normal(mu_norm, sigma_norm, 1000)

# Построение гистограммы
plt.hist(data_norm, bins=30, density=True, alpha=0.6, color='m')

plt.xlabel('Значение случайной величины')
plt.ylabel('Плотность вероятности')
plt.title('Гистограмма нормального распределения')
plt.grid(True)
plt.show()

# Параметры нормального распределения
mu_norm = 0
sigma_norm = 1

# Эмпирические значения
mean_empirical_norm = np.mean(data_norm)
var_empirical_norm = np.var(data_norm)

# Теоретические значения
mean_theoretical_norm = mu_norm
var_theoretical_norm = sigma_norm ** 2

print("Нормальное распределение:")
print("Теоретическое среднее:", mean_theoretical_norm)
print("Эмпирическое среднее:", mean_empirical_norm)
print("Теоретическая дисперсия:", var_theoretical_norm)
print("Эмпирическая дисперсия:", var_empirical_norm)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Параметры бета-распределения
a_beta = 2.0
b_beta = 5.0

# Генерация выборки из бета-распределения
data_beta = beta.rvs(a_beta, b_beta, size=1000)

# Построение гистограммы
plt.hist(data_beta, bins=30, density=True, alpha=0.6, color='m')

# Построение теоретической плотности распределения
x = np.linspace(0, 1, 100)
plt.plot(x, beta.pdf(x, a_beta, b_beta), 'r-', lw=2, label='theoretical pdf')

plt.xlabel('Значение случайной величины')
plt.ylabel('Плотность вероятности')
plt.title('Гистограмма и теоретическая плотность бета-распределения')
plt.legend()
plt.grid(True)
plt.show()

# Эмпирические значения
mean_empirical_beta = np.mean(data_beta)
var_empirical_beta = np.var(data_beta)

# Теоретические значения
mean_theoretical_beta = a_beta / (a_beta + b_beta)
var_theoretical_beta = (a_beta * b_beta) / ((a_beta + b_beta) ** 2 * (a_beta + b_beta + 1))

print("Бета-распределение:")
print("Теоретическое среднее:", mean_theoretical_beta)
print("Эмпирическое среднее:", mean_empirical_beta)
print("Теоретическая дисперсия:", var_theoretical_beta)
print("Эмпирическая дисперсия:", var_empirical_beta)
```
