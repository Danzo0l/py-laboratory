```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Функция для вычисления количества отличающихся пикселей между двумя изображениями
def count_different_pixels(file1, file2):
    img1 = np.array(Image.open(file1))
    img2 = np.array(Image.open(file2))

    # Подсчет отличающихся пикселей
    different_pixels = np.sum(img1 != img2)

    return different_pixels

if __name__ == '__main__':
    original_file = 'C:\\Users\\dryag\\Documents\\cryptography\\cast_128\\input.bmp'
    encrypted_file = 'C:\\Users\\dryag\\Documents\\cryptography\\cast_128\\img_d.bmp'

    different_pixels = count_different_pixels(original_file, encrypted_file)

    print(f"Number of different pixels: {different_pixels}")
    
    img1 = Image.open(original_file)
    img2 = Image.open(encrypted_file)

    # Приблизить оба изображения
    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.xlim(300, 500)
    plt.ylim(450, 250)
    plt.title('Source image')

    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.xlim(300, 500)
    plt.ylim(450, 250)
    plt.title('Decrypted image')

    plt.show()
```
